***
보스턴 집값 예측하기
***

```python
# Scikit-learn 라이브러리에서 제공하는 보스턴 집값 데이터를 불러온다.
raw_data = datasets.load_boston()
raw_data
```

```python
# Scikit-learn에서 제공하는 데이터는 data(피쳐)부분과 target(레이블, 클래스, 실제값)부분으로 구성되어있다.
# 'feature_names'(피쳐 이름), 'DESCR'(피쳐의 설명), 'filename'(csv파일 이름), 'data_module'(데이터 모듈이름) 정보를 제공한다.
```

```python
# 피쳐부분을 추출해 데이터프레임으로 만든다.
x_train = pd.DataFrame(raw_data.data)
print(x_train.shape)
x_train.head()
```

```python
# 타겟부분을 추출해서 데이터프레임을 만든다.
y_train = pd.DataFrame(raw_data.target)
print(y_train.shape)
y_train.head()
```

```python
# 피쳐와 레이블 데이터를 결합해서 전체 프레임을 생성한다.
# pandas의 concat() 메소드를 이용해서 데이터프레임을 합칠수 있다.
#concat() 메소드의 axis 속성으로 데이터프레임의 결합 방향을 지정할 수 있다.
# axis속성을 생략하면 0이 기본값이 데이터프레임을 위에서 아래방향으로 합치고, 1이면 왼쪽에서 오른쪼긍로 합친다.
raw_data_df = pd.concat([x_train, y_train], axis=1)
print(raw_data_df.shape)
raw_data_df.head()
```

```python
# 피쳐와 레이블을 결합한 데이터프레임의 열 이름을 지정한다.
feature_names = raw_data.feature_names
print(type(feature_names))
print(feature_names)
columns = np.append(feature_names, ['LABEL'])
print(columns)
raw_data_df.columns = columns
raw_data_df
```

```python
# 필요하다면 작업이 완료된 데이터프레임을 csv파일로 저장한다.
raw_data_df.to_csv(f'./data/{raw_data.filename}', index=False)
```
